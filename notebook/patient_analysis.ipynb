# 🏥 AI-Based Patient Progress Analysis
This notebook analyzes health data of 20 patients and provides insights and classification based on vital signs.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
# Load the dataset
df = pd.read_csv('../data/patients_data.csv')
df.head()
```

```python
# Summary statistics
df.describe(include='all')
```

```python
# Distribution of AI Status
sns.countplot(data=df, x='AI_Status')
plt.title('Distribution of AI Health Status')
plt.show()
```

```python
# Average vitals by AI_Status
df['Systolic'] = df['Blood_Pressure'].apply(lambda x: int(x.split('/')[0]))
df['Diastolic'] = df['Blood_Pressure'].apply(lambda x: int(x.split('/')[1]))
avg_vitals = df.groupby('AI_Status')[['Heart_Rate', 'Oxygen_Level', 'Temperature', 'Systolic', 'Diastolic']].mean()
avg_vitals.plot(kind='bar', figsize=(10,6), title='Average Vitals by AI Status')
plt.ylabel('Average Value')
plt.xticks(rotation=0)
plt.show()
```
